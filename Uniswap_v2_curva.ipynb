```python
import matplotlib.pyplot as plt
import numpy as np

# A constante do pool de liquidez (Uniswap V2)
k = 40000000

# Pontos P e Q (dados fornecidos)
eth_q, usdc_q = 80, 500000
eth_p, usdc_p = 100, 400000

# Gerar uma série de pontos para desenhar a curva suavemente
# Usando um intervalo um pouco maior para a curva
eth_valores = np.linspace(75, 105, 300)
usdc_valores = k / eth_valores

# Configurar o estilo do gráfico
plt.style.use('seaborn-v0_8-whitegrid') # Um estilo de grade sutil

plt.figure(figsize=(15, 7), facecolor='white') # Aumenta o tamanho e deixa o fundo branco

# Plotar a curva com uma linha mais grossa e cor personalizada
plt.plot(eth_valores, usdc_valores, color='#0A79DF', linewidth=3)

# Adicionar um preenchimento (sombra) abaixo da curva para um efeito de profundidade
plt.fill_between(eth_valores, usdc_valores, color='#0A79DF', alpha=0.15)

# Plotar os pontos P e Q com um marcador mais visível e borda
plt.plot([eth_q, eth_p], [usdc_q, usdc_p], 'o', color='#FF6347', markersize=10, markeredgecolor='black', zorder=5)

# Adicionar as linhas de projeção pontilhadas (mais finas e cinzas)
plt.vlines([eth_q, eth_p], ymin=plt.ylim()[0], ymax=[usdc_q, usdc_p], color='gray', linestyle='--', linewidth=1, alpha=0.7)
plt.hlines([usdc_q, usdc_p], xmin=plt.xlim()[0], xmax=[eth_q, eth_p], color='gray', linestyle='--', linewidth=1, alpha=0.7)

# Adicionar anotações para os pontos P e Q
# Usando 'bbox' para um fundo e 'ha' para alinhamento
plt.annotate('Q', (eth_q, usdc_q), textcoords="offset points", xytext=(-15, 15), ha='center',
             fontsize=14, color='#333333', weight='bold',
             bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8, ec='gray', lw=0.5))
plt.annotate('P', (eth_p, usdc_p), textcoords="offset points", xytext=(15, 15), ha='center',
             fontsize=14, color='#333333', weight='bold',
             bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.8, ec='gray', lw=0.5))

# Configurar os eixos e o título
plt.title('Balanço de Ativos em um Pool de Liquidez (Uniswap V2)', fontsize=16, pad=20, weight='bold')
plt.xlabel('Saldo de ETH', fontsize=12, labelpad=15)
plt.ylabel('Saldo de USDC', fontsize=12, labelpad=15)

# Definir os ticks dos eixos manualmente para destacar os pontos
plt.xticks([80, 100])
plt.yticks([400000, 500000])

# Ajustar os limites dos eixos para uma visualização melhor
plt.xlim(75, 105)
plt.ylim(370000, 530000)

# Remover as bordas superior e direita do gráfico para um visual mais limpo
ax = plt.gca()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.5)
ax.xaxis.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
