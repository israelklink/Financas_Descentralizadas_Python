```python
import matplotlib.pyplot as plt
```

```python

# Inicializar o pool
reserva_A = 1000  # Ex: ETH
reserva_B = 1000  # Ex: USDC
k = reserva_A * reserva_B

print("k - produto constante:",k)

precos = []
valores_trocados = []

for dx in range(1, 300, 5):  # quantidade de Token A a ser trocado
    nova_A = reserva_A + dx
    nova_B = k / nova_A
    dy = reserva_B - nova_B
    preco_efetivo = dy / dx

    precos.append(preco_efetivo)
    valores_trocados.append(dx)

plt.plot(valores_trocados, precos)
plt.xlabel("Token A trocado")
plt.ylabel("Pre√ßo efetivo (Token B por Token A)")
plt.title("Slippage no AMM (x * y = k)")
plt.grid(True)
plt.show()
```
