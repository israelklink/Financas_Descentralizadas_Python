```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a quantidade inicial de tokens no pool
x0 = 10        # Token X (por exemplo, ETH)
y0 = 20000     # Token Y (por exemplo, USDC)
k = x0 * y0    # Constante invariante

# Gera valores para x no intervalo (evita divisão por zero)
x = np.linspace(0.1, 2 * x0, 1000)

# Calcula y a partir da curva x * y = k
y = k / x

# Preço instantâneo: y/x
preco = y / x

# Plot da curva AMM
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Curva AMM: $x \cdot y = k$', color='blue')
plt.title('Curva do AMM da Uniswap v2')
plt.xlabel('Token X (ex: ETH)')
plt.ylabel('Token Y (ex: USDC)')
plt.grid(True)
plt.legend()
plt.axhline(y=y0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(x=x0, color='gray', linestyle='--', linewidth=0.5)
plt.show()

# Plot do preço instantâneo
plt.figure(figsize=(10, 6))
plt.plot(x, preco, label='Preço Instantâneo: y / x', color='green')
plt.title('Preço Instantâneo em Função da Quantidade de Token X')
plt.xlabel('Token X (ex: ETH)')
plt.ylabel('Preço (USDC por ETH)')
plt.grid(True)
plt.legend()
plt.show()
```
